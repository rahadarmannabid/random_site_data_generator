```python
from datetime import datetime, timedelta
import time
import pandas as pd
import openpyxl
from random import randint
```

```python
district=['Dhaka','Rangpur', 'Nilphamari', 'Rajshahi', 'Cumilla']
tower_id=['36510671', ' 102111034921', ' 102111034621', ' 1029321034921', '76510671']
initial_reading= ['23023', '13029', '33023','3003', '73023']
latitude=['23.7816','22.3521', '25.9363','24.3745','23.4607']
longitude=['90.4165','71.7178','88.8407', '88.6042', '91.1809'] 
data = []
for j in range(0,5):
    now=datetime.now()
    commutative_reading = int(initial_reading[j])
    for  i in range(0,1500):
        now=now - timedelta(hours = 1)
        #date_time = now.strftime("%m/%d/%Y_%H")
        only_date = now.strftime("%m/%d/%Y")
        only_time = now.strftime("%H:%M:%S")
        date_and_time = only_date + " " + only_time
        hourly_consumption = randint(0, 3)
        commutative_reading = commutative_reading + hourly_consumption
        data.append([district[j],  tower_id[j], only_date, only_time, date_and_time, commutative_reading, hourly_consumption, latitude[j], longitude[j]])
    
    
```

```python
df = pd.DataFrame(data, columns=['district',  'tower_id', 'only_date', 'only_time' , 'date_and_time' , 'commutative_reading', 'hourly_consumption', 'laditude', 'longitude'])
df
```

```python
df.to_excel('ai_meter_managment.xlsx', sheet_name='all_data')
```
